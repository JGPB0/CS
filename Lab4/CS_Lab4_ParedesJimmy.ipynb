###2. Write a Python program that uses the hashlib library to apply the hash algorithms MD5, SHA-1, SHA-256, SHA-384, and SHA-512 to a given text string.

````python
import hashlib

def apply_hash_algorithms(text):
    # MD5
    md5_hash = hashlib.md5(text.encode()).hexdigest()

    # SHA-1
    sha1_hash = hashlib.sha1(text.encode()).hexdigest()

    # SHA-256
    sha256_hash = hashlib.sha256(text.encode()).hexdigest()

    # SHA-384
    sha384_hash = hashlib.sha384(text.encode()).hexdigest()

    # SHA-512
    sha512_hash = hashlib.sha512(text.encode()).hexdigest()

    return md5_hash, sha1_hash, sha256_hash, sha384_hash, sha512_hash

# Example text
text = "Hello, world!"

# Apply hash algorithms
md5_hash, sha1_hash, sha256_hash, sha384_hash, sha512_hash = apply_hash_algorithms(text)

# Print results
print("MD5 hash:", md5_hash)
print("SHA-1 hash:", sha1_hash)
print("SHA-256 hash:", sha256_hash)
print("SHA-384 hash:", sha384_hash)
print("SHA-512 hash:", sha512_hash)
````
